```python
train = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
test = pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
```

```python
train.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
```

```python
train['HomePlanet'] = label_encode.fit_transform(train['HomePlanet'])
train['CryoSleep'] = label_encode.fit_transform(train['CryoSleep'])
train['Destination'] = label_encode.fit_transform(train['Destination'])
train['VIP'] = label_encode.fit_transform(train['VIP'])
train['Name'] = label_encode.fit_transform(train['Name'])
train['Cabin'] = label_encode.fit_transform(train['Cabin'])
```

```python
train.isnull().sum()
```

```python
train['Age'].fillna(value=0, inplace = True)
train['RoomService'].fillna(value=0, inplace = True)
train['FoodCourt'].fillna(value=0, inplace = True)
train['ShoppingMall'].fillna(value=0, inplace = True)
train['Spa'].fillna(value=0, inplace = True)
train['VRDeck'].fillna(value = 0, inplace = True)
```

```python
train.isnull().sum()
```

```python
x_train = train.drop(['PassengerId','Transported'],axis = 1)
y_train = train['Transported']
```

```python
x_train
```

```python
test.head(5)
```

```python
test['HomePlanet'] = label_encode.fit_transform(test['HomePlanet'])
test['CryoSleep'] = label_encode.fit_transform(test['CryoSleep'])
test['Destination'] = label_encode.fit_transform(test['Destination'])
test['VIP'] = label_encode.fit_transform(test['VIP'])
test['Name'] = label_encode.fit_transform(test['Name'])
test['Cabin'] = label_encode.fit_transform(test['Cabin'])
```

```python
test['Age'].fillna(value=0, inplace = True)
test['RoomService'].fillna(value=0, inplace = True)
test['FoodCourt'].fillna(value=0, inplace = True)
test['ShoppingMall'].fillna(value=0, inplace = True)
test['Spa'].fillna(value=0, inplace = True)
test['VRDeck'].fillna(value = 0, inplace = True)
```

```python
x_test = test.drop(['PassengerId',],axis = 1)
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
regressor.fit(x_train,y_train)
predictions = regressor.predict(x_test)

# Define threshold
threshold = 0.5

# Apply threshold and convert to Boolean values
boolean_array = predictions >= threshold

# Print the Boolean array
print(boolean_array)
```

```python
boolean_array
```

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Transported': boolean_array})
output.to_csv('submission.csv', index=False)
```
